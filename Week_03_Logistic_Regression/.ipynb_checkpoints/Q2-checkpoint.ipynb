```python
def mapFeature(x1,x2,degree):
    out = np.ones(len(x1)).reshape(len(x1),1)
    out= np.c_[out,X1,X2]
    for i in range(1,degree+1):
        for j in range(i+1):
            terms= (x1**(i-j) * x2**j)
            out= np.c_[out,terms]
    return out
```

```python
def costFunctionReg(w,X,y,lambda_):
        # print("costFunctionReg")
        y = np.round(np.array(y))
        m,col = X.shape
        x = X.T
        
        h = sigmoid(np.dot(w,x))
        
        J = (-1*np.dot(y,np.log(h)) - np.dot(1-y,np.log(1-h)))/m + lambda_*np.sum(np.square(w))/(2*m) # Cost 'J' should be a scalar
        
        
        grad =  (h - y)
        grad_J = np.dot(x,grad.T)
        
        grad =  np.divide(grad_J,m)    +  lambda_*w/(m)   # Gradient 'grad' should be a vector
        print(J)
        return J, grad
```

```python
def sigmoid(z):
        
        return 1/(1 + np.exp(-z))
```

```python
def predict(w,X):
        # print("predict")
        p = np.round(sigmoid(np.dot(w,X)))           # 'p' should be a vector of size equal to that of vector 'y'
        print(np.sum(p))
        return p
```

```python
def minCostFun( w_ini, X_train, y_train, iters,lambda_):
        row,col = X_train.shape
        m = col
        options = {'maxiter':iters}

        res = optimize.minimize(costFunctionReg,w_ini,(X_train,y_train,lambda_),jac = True,method='TNC',options = options)
        
        cost = res.fun
        
        w_opt =  res.x     # Optimized weights rounded off to 3 decimal places
        return w_opt
```

```python
X1 = np.array(dat[:,0])
X2 = np.array(dat[:,1])

print(type(X1))
Y = dat[:,2]

m = X1.size
print(m)
lambda_ =100
# X_set  = np.vstack((np.ones(m),X1,X2,np.square(X1),np.multiply(X1,X2),np.square(X2)))
X_set = mapFeature(X1,X2,6).T
print(X_set.shape)
w = minCostFun(np.zeros(28),X_set.T,Y,1000,lambda_)
```

```python
u = np.linspace(-1, 1.5, 50) # 1D array from -1 to 1.5 based on the limits of data
v = np.linspace(-1, 1.5, 50)

z = np.zeros((u.size, v.size))
# Evaluate z = w*x over the grid
for i, ui in enumerate(u):
    for j, vj in enumerate(v):
#         z[i, j] = np.dot(np.array([1,ui, vj,ui*ui,ui*vj,vj*vj]), w)
        z[i, j] = np.dot(mapFeature(np.array([ui]),np.array([vj]),6), w)

z = z.T
    
plt.contour(u, v, z, levels=[0], linewidths=2, colors='g') # Plots contour lines
plt.contourf(u, v, z, levels=[np.min(z), 0, np.max(z)], cmap='Greens', alpha=0.4) # Plots filled contours

plt.plot(data_zero[:,0],data_zero[:,1],'o', label = 'Y = 1')
plt.plot(data_one[:,0],data_one[:,1],'x', label = 'Y = 0')


# Specified in plot order
    
plt.xlabel('Microchip Test 1')
plt.ylabel('Microchip Test 2')
plt.legend()
plt.title('lambda = %0.3f' % lambda_)
plt.savefig("lambda100",dpi=500)
```
