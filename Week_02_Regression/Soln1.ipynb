```python
import os
import matplotlib.pyplot as plt
import numpy as np
from Week2_Q1_Template import nlr
plt.style.use('ggplot')
```

```python
data = np.loadtxt('prob1data.txt', delimiter=',')
```

```python
x = data[0]
y = data[1]
mat = np.vstack([np.ones(x.size),x,np.square(x)])
plt.plot(x,y,'o')
plt.title("Training data")
plt.xlabel('Time')
plt.ylabel('Height')  
plt.savefig('TrainingData_1.png',dpi=500)
```

```python
iters = 50000

w = np.random.rand(3)
nlr_obj = nlr()


b = nlr_obj.SgradientDescent(mat.T,y,w.copy(),0.01,iters)
a = nlr_obj.BgradientDescent(mat.T,y,w.copy(),0.01,iters)

x_ar = [i for i in range(iters)]
y_ar = a[1]
y_ar_s = b[1]



plt.plot(x_ar,y_ar_s,label='Stochastic gradient descent')
plt.plot(x_ar,y_ar,label='Batch gradient descent')
plt.title("J vs iterations")
plt.xlabel('No of iterations')
plt.ylabel('J')  
plt.legend()
plt.savefig('GradDescent_1_copy.png',dpi=500)
plt.show()
```

```python
x_arr = np.linspace(0,2.5,100)
# a = [[ 1.15839522, 11.69078221, -4.37995635],[0]]
a = [[-0.13573482 ,14.20036254, -5.04168892],[0]]
y_arr = a[0][0] + a[0][1]*x_arr + a[0][2]*(x_arr**2)

plt.plot(x,y,'o',label = "Training Data")
plt.plot(x_arr,y_arr,label = "Prediction")
plt.title("Prediction")
plt.xlabel('Time')
plt.ylabel('Height')  
plt.legend()
plt.savefig('Prediction_para_1.png',dpi=500)
plt.show()
```

```python
iters = 10000

a = nlr_obj.BgradientDescent(x,y,w.copy(),0.1,iters)
b = nlr_obj.BgradientDescent(x,y,w.copy(),0.05,iters)
c = nlr_obj.BgradientDescent(x,y,w.copy(),0.01,iters)
d = nlr_obj.BgradientDescent(x,y,w.copy(),0.005,iters)

x_ar = [i for i in range(iters)]
y_ar_a = a[1]
y_ar_b = b[1]
y_ar_c = c[1]
y_ar_d = d[1]

plt.plot(x_ar,y_ar_a,label=" = 0.1")
plt.plot(x_ar,y_ar_b,label=" = 0.05")
plt.plot(x_ar,y_ar_c,label=" = 0.01")
plt.plot(x_ar,y_ar_d,label=" = 0.005")
plt.title("Variaton of J with alpha")
plt.xlabel('No of iterations')
plt.ylabel('J')  
plt.legend()
plt.savefig('Prediction_1_blowup.png',dpi=1000)
plt.show()
```
